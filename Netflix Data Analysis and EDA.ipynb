Netflix Data EDA

```
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Data Loading

df=pd.read_csv(r'netflix_titles.csv')


df=df.drop(columns=["show_id"])
df


df.isna().sum()

# Data Cleaning

df["director"]=df["director"].replace(np.nan,"Unknown")
df

df["cast"]=df["cast"].replace(np.nan,"Unknown")
df["country"]=df["country"].replace(np.nan,"Unknown")
df

df.isna().sum()
df=df.dropna()
df

df.isna().sum()

df[["digit","time"]]=df["duration"].str.split(" ",expand=True)
df["digit"]=df["digit"].astype(int)
df["time"]=df["time"].map({"min":1,"Seasons":225,"Season":225})
df.dtypes

df=df.drop(columns=['duration'])
df["duration"]=df["digit"]*df["time"]

df["month"]=df["date_added"].replace(r'[^A-Za-z]','',regex=True)




```

```
df1=df.groupby("release_year")["duration"].aggregate("count")
df1=df1.reset_index()
df1
```

#Exploratory Data Analysis

1.) counting no.of movies released in olden days


```
df11=df1.head(10)

plt.bar(df11["release_year"],df11["duration"])
plt.plot(df11["release_year"],df11["duration"])
plt.show()
```

2.) counting movies released in recent days


```
df12=df1.tail(10)

plt.bar(df12["release_year"],df12["duration"])
plt.show()
```

3)Top 10 movies with highest duration

```
df2=df[df["type"]=="Movie"]
df2=df2.sort_values(by="duration",ascending=False)
df2=df2.head(10)
plt.figure(figsize=(25,7))
sns.histplot(x=df2['title'],y=df2["duration"],color="red")
plt.show()
```

4.) bottom 10 Movies with lowest
 duration


```
df2=df[df["type"]=="Movie"]
df2=df2.sort_values(by="duration",ascending=True)
df2=df2.tail(10)
plt.figure(figsize=(30,7))
plt.plot(df2['title'],df2["duration"],color="blue",marker="o")
plt.show()
```

5)visualizing which type of movies or tv shows....etc people are interested in


```
# grp by
# descending
# visualize
df3=df.groupby("type")["duration"].aggregate("count")
df3=df3.reset_index()
plt.figure(figsize=(25,10))
plt.bar(df3["type"],df3["duration"])
plt.show()
```

6)Visualizing in which country ,most no.of movies are releasing

```
from itertools import filterfalse
df4=df[df["country"]!="Unknown"]
df4=df4[df4["type"]=="Movie"]
df4["country"]=df4["country"].str.split(",",expand=True)[0]
df4

df4=df4.groupby("country")["duration"].aggregate("count")
df4=df4.reset_index()
df4=df4.sort_values(by="duration",ascending=False)
df4=df4.head(10)
df4

plt.figure(figsize=(15,7))
plt.pie(df4["duration"],labels=df4["country"],autopct="%0.0f")
plt.show()
```

7)visualizing the director who made most no.of movies

```
df5=df[df["director"]!="Unknown"]
df5=df5.groupby("director")["duration"].aggregate("count")
df5=df5.reset_index()
df5=df5.sort_values(by="duration",ascending=False)
df5=df5.head(10)
df5.columns=df5.columns.str.replace("duration","count_of_movies")
plt.figure(figsize=(25,10))
plt.bar(df5["director"],df5["count_of_movies"])
plt.show()
```

8)top 10 kind of movies and tv shows

```
df6=df
df6["listed_in"]=df6["listed_in"].str.split(",",expand=True)[0]
df6=df6.groupby("listed_in")["duration"].aggregate("count")

df6=df6.reset_index()
df6=df6.sort_values(by="duration",ascending=False)
df6=df6.head()
df6

plt.figure(figsize=(10,8))
plt.pie(df6["duration"], labels=df6["listed_in"], autopct="%0.0f%%")
plt.show()
```

```
df["month"]=df["date_added"].replace(r'[^A-Za-z]','',regex=True)
df7=df.groupby("month")["release_year"].aggregate("count")
df7=df7.reset_index()
df7=df7.sort_values(by="release_year",ascending=False)
df7

plt.pie(df7["release_year"],labels=df7["month"],autopct="%0.0f%%")
plt.show()
```
